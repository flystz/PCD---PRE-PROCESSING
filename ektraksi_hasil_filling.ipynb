```python
import cv2
import numpy as np
from skimage.measure import label, regionprops
import matplotlib.pyplot as plt

# Menggantikan 'nama_file_gambar' dengan nama file gambar yang ingin diproses
nama_file_gambar = 'ss an filing/gambar 1.jpg'

# Membaca gambar dalam mode grayscale
gambar = cv2.imread(nama_file_gambar, cv2.IMREAD_GRAYSCALE)

# Binarisasi gambar
_, binary = cv2.threshold(gambar, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Menghubungkan komponen yang terhubung dalam gambar biner
labeled_image = label(binary)

# Mengambil properti dari setiap komponen terhubung
regions = regionprops(labeled_image)

# Looping untuk menghitung luas, keliling, dan eksentrisitas setiap komponen
for region in regions:
    area = region.area
    perimeter = region.perimeter
    eccentricity = region.eccentricity
    metric = 4 * 3.14 * area / (perimeter ** 2)


    print('Objek', region.label)
    print('Luas:', area)
    print('Keliling:', perimeter)
    print('Eksentrisitas:', eccentricity)
    print('Metric:', metric)
    print('-----------------')

    # Menampilkan citra asli dengan bounding box pada setiap objek
    min_row, min_col, max_row, max_col = region.bbox
    cv2.rectangle(gambar, (min_col, min_row), (max_col, max_row), (0, 255, 0), 2)

# Menampilkan citra asli dan objek yang di-highlight dengan bounding box
fig, ax = plt.subplots()
ax.imshow(gambar, cmap='gray')
plt.axis('off')  # Menghilangkan sumbu koordinat

plt.show()
```
