```python
import cv2
import numpy as np
from skimage.measure import label, regionprops
import matplotlib.pyplot as plt

# Baca gambar
gambar = cv2.imread('gambar 1.jpg')

# Segmentasi menggunakan deteksi warna
hsv = cv2.cvtColor(gambar, cv2.COLOR_BGR2HSV)
lower = np.array([0, 0, 170])  # Rentang warna bawah (misal: putih)
upper = np.array([180, 25, 255])  # Rentang warna atas (misal: putih)
mask = cv2.inRange(hsv, lower, upper)
segmentasi = cv2.bitwise_and(gambar, gambar, mask=mask)

# Erosi
kernel = np.ones((7, 7), np.uint8)
erosi = cv2.erode(mask, kernel, iterations=1)

# Dilasi
dilasi = cv2.dilate(erosi, kernel, iterations=1)

# Opening (erosi diikuti dilasi)
opening = cv2.morphologyEx(erosi, cv2.MORPH_OPEN, kernel)

# Closing (dilasi diikuti erosi)
closing = cv2.morphologyEx(dilasi, cv2.MORPH_CLOSE, kernel)

# Filter Median
median = cv2.medianBlur(closing, 3)

# Filling (Flood fill)
filled = np.copy(median)
h, w = filled.shape
mask_floodfill = np.zeros((h + 2, w + 2), np.uint8)
for row in range(h):
    for col in range(w):
        if filled[row, col] == 0:
            cv2.floodFill(filled, mask_floodfill, (col, row),2)

# Menampilkan gambar asli dan hasil dalam satu jendela
fig, ax = plt.subplots(2, 4, figsize=(12, 6))

ax[0, 0].imshow(cv2.cvtColor(gambar, cv2.COLOR_BGR2RGB))
ax[0, 0].set_title('Gambar Asli')
ax[0, 0].axis('off')

ax[0, 1].imshow(mask, cmap='gray')
ax[0, 1].set_title('Segmentasi (Deteksi Warna)')
ax[0, 1].axis('off')

ax[0, 2].imshow(erosi, cmap='gray')
ax[0, 2].set_title('Erosi')
ax[0, 2].axis('off')

ax[0, 3].imshow(dilasi, cmap='gray')
ax[0, 3].set_title('Dilasi')
ax[0, 3].axis('off')

ax[1, 0].imshow(opening, cmap='gray')
ax[1, 0].set_title('Opening')
ax[1, 0].axis('off')

ax[1, 1].imshow(closing, cmap='gray')
ax[1, 1].set_title('Closing')
ax[1, 1].axis('off')

ax[1, 2].imshow(median, cmap='gray')
ax[1, 2].set_title('Filter Median')
ax[1, 2].axis('off')

ax[1, 3].imshow(filled, cmap='gray')
ax[1, 3].set_title('Filling')
ax[1, 3].axis('off')

plt.tight_layout()
plt.show()
```
